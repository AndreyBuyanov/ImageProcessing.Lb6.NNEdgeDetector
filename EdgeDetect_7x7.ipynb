```python
from EdgeDetector import EdgeDetector
```

```python
from os import listdir
from os.path import isfile, join
```

```python
def get_file_list(path):
    return [join(path, f) for f in listdir(path) if isfile(join(path, f))]
```

```python
data_directory = '102flowers'
```

```python
file_list = get_file_list(path=data_directory)
train_file_list = file_list[:30]
test_file_list = file_list[30:35]
```

```python
kernel_size = 7
```

```python
edge_detector = EdgeDetector(
    train_file_list=train_file_list,
    test_file_list=test_file_list,
    kernel_size=kernel_size)
```

```python
import cv2
```

```python
image = cv2.imread(file_list[36])
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edge_image = cv2.Canny(image=gray_image, threshold1=100, threshold2=200)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(16,16))
plt.imshow(gray_image.astype('uint8'), cmap=plt.get_cmap('gray'))
plt.show()
```

```python
plt.figure(figsize=(16,16))
plt.imshow(edge_image.astype('uint8'), cmap=plt.get_cmap('gray'))
plt.show()
```

```python
result = edge_detector.detect(input_image=gray_image)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(result.astype('uint8'), cmap=plt.get_cmap('gray'))
plt.show()
```
